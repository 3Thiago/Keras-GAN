```
import os
from packaging import version
from PIL import Image

pasta = '/content/images/apple2orange'
filenames =  os.listdir(pasta)
print(filenames)


images = []
for filename in filenames:
  a = pasta+'/'+filename
  images.append(a)
  print(images)



im1 = Image.open('/content/images/apple2orange/0_0.png')

if version.parse(Image.PILLOW_VERSION) < version.parse("3.4"):
    print("Pillow in version not supporting making animated gifs")
    print("you need to upgrade library version")
    print("see release notes in")
    print("https://pillow.readthedocs.io/en/latest/releasenotes/3.4.0.html#append-images-to-gif")
else:
    im1.save("/content/images/out.gif", save_all=True, append_images=images, duration=100, loop=0)
```

```
import glob
import moviepy.editor as mpy

gif_name = '/content/images/outputName'
fps = 12
file_list = glob.glob('/content/images/apple2orange/*.png') # Get all the pngs in the current directory
list.sort(file_list, key=lambda x: int(x.split('_')[1].split('.png')[0])) # Sort the images by #, this may need to be tweaked for your use case
clip = mpy.ImageSequenceClip(file_list, fps=fps)
clip.write_gif('{}.gif'.format(gif_name), fps=fps)
```
